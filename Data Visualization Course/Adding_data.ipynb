```python
FileName_GDP = 'World_GDP.csv'
df_GDP = pd.read_csv(FileName_GDP, low_memory = False)
df_GDP.head()
```

```python
df_GDP_new = pd.melt(df_GDP, 
        id_vars='Country Name', 
        value_vars=list(df_GDP.columns[1:]), # list of days of the week
        var_name='Year', 
        value_name='GDP in $')
```

```python
df_GDP_new.head()
df_GDP_new.to_csv("GDP_new.csv")
```

```python
df_GDP_new["Country_year"] = df_GDP_new["Country Name"] + str(df_GDP_new["Year"])
df_GDP_new.head()
```

```python
FileName3 = 'GDP_new.csv'
df_gdp = pd.read_csv(FileName3, low_memory = False)
```
