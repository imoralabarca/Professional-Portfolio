```python
import pandas as pd
import numpy as np
import datetime
```

```python
FileName = 'globalterrorismdb_0718dist.csv'
df = pd.read_csv(FileName, low_memory = False)
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None

df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region','attacktype1_txt':'Attack_type','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive', 'nperps':'Number_attackers', 'crit1': 'PERS_attack', 'crit2':'Larg_aud_attack', 'crit3':'Out_IHL_attack','natlty1_txt' : 'Target_nationality', 'targsubtype1_txt': 'Target_subtype'},inplace=True)

         
columns = ['eventid', 'Year', 'Month','Day', 'approxdate', 'Country','Region', 'provstate', 'city', 'latitude','longitude', 'multiple', 'related', 'PERS_attack','Larg_aud_attack', 'Out_IHL_attack', 'suicide', 'Attack_type','Target_type', 'Target_subtype', 'Target_nationality', 'Group', 'guncertain1', 'individual', 'Number_attackers', 'nperpcap', 
            'Weapon_type', 'weapsubtype1_txt', 'Killed', "nkillter", 'Wounded', 'property', 'propextent', 'propextent_txt', 'propvalue', 'ishostkid','ransom', 'INT_LOG', 'INT_IDEO','INT_MISC','INT_ANY']

DV = df[columns]
DV.head()
```

```python
# get names of indexes for which
# column Month has value 0
index_names = DV[ DV['Month'] == 0 ].index
  
# drop these row indexes
# from dataFrame
DV.drop(index_names, inplace = True)
DV[DV["Month"] ==0].count() #check that we don't have any remaining row with month = 0
```

```python
DV['dateInt'] = DV['Year'].astype(str) + DV['Month'].astype(str).str.zfill(2)+ DV['Day'].astype(str).str.zfill(2)
DV['Date'] = pd.to_datetime(DV['dateInt'], format='%Y%m%d')
DV.head()
```

```python
DV['Number_attackers'].replace(-99, np.nan, inplace = True)
DV['nperpcap'].replace(-99, np.nan, inplace = True)
DV['ishostkid'].replace(-9, 'Unknown', inplace = True)
DV['ishostkid'].replace(np.nan, 'Unknown', inplace = True)
DV['property'].replace(-9, 'Unknown', inplace = True)
DV['ransom'].replace(-9, 'Unknown', inplace = True)
DV['ransom'].replace(np.nan, 'Unknown', inplace = True)
DV['INT_LOG'].replace(-9, 'Unknown', inplace = True)
DV['INT_IDEO'].replace(-9, 'Unknown', inplace = True)
DV['INT_MISC'].replace(-9,'Unknown', inplace = True)
DV['INT_ANY'].replace(-9, 'Unknown', inplace = True)
DV.head()
```

```python
DV['Casualties'] = DV['Killed'] + DV['Wounded']
DV.head()
```

```python
DV.to_csv('Terrorism_df_V4.csv')
```
